```python
from tkinter import *
import pandas as pd
import random as rd
import datetime as dt
import math
import re

BG_GRAY = "#ABB2B9"
BG_COLOR = "#17202A"
TEXT_COLOR = "#EAECEE"

FONT = "Helvetica 14"
FONT_BOLD = "Helvetica 13 bold"

CHATBOT_NAME = "Chatbot Eternal"

count = 1
ordering_progress = 1
ordering = False

dataSet = pd.read_excel('conversationSet.xlsx')#read dataset from excel file
dataSet['Question'] = dataSet['Question'].astype(str)#convert the question column value to string

#user information
userName = "User"#username
userAddress = ""
userPhoneNo = ""
validatingPhoneNo = 1
userInformation = False
request = ""

userInput = ""#user input
exit = False#if exit = True then will end the chat

#detect key value- to get username
greetings = ["MY NAME IS ", "I'M ", "I AM "]

#greeting value
greetingValue = ['Hi', 'Greetings', 'Hey', 'Hello', 'Nice To Meet You']

#features
features = ["ALL MENU","FOOD LIST","BEVERAGE LIST","SET LIST","HOW TO ORDER","HELP","FREQUENTLY ASKED QUESTIONS","RECOMMENDED FOOD","RECOMMENDED BEVERAGE","RECOMMENDED SET"]

#food price
foodPrice = [12.90,17.90,28.90,21.90,29.90,29.90,27.90,29.90,28.90,30.90]
beveragePrice = [19.90,18.90,12.90,10.90,19.90,18.90]
packagePrice = [27.90,28.90,31.90,31.90,30.90,31.90,31.90,31.90,25.90]
foodMenu = ['Creamy Mushroom Soup','Tuna Croissant','Chicken Chop','Fish & chips','Salmon Steak',
            'Spaghetti Aglio e olio','Chicken Lasagne','Creamy Bacon Carbonara','Seafood Carbonara',
            'Seafood Spaghetti','BBQ Steak']
beverageMenu = ['Cappuccino','Americano','Orange Juice Shake','Mix Fruit Shake','Tiramisu Milkshake','Oreo Shake']
packageMenu = ['American style Breakfast & Americano','British Style Breakfast & Cappuccino',
               'Creamy Bacob Carbonara & Apple Tea','Spaghetti Aglio E Olio & Apple Tea',
               'Chicken Chop & Orange Juice','Seafood Cabonara & Orange Juice',
               'Tuna Croissant+Creamy Mushroom Soup & Green Tea','BBQ Steak & Mix Fruit Shake',
               'Fish & Chips & Mix Fruit Shake']

#lucky draw
luckyDraw = ['DISCOUNT10P','DISCOUNT15P','DISCOUNT30P','DISCOUNT50P','DISCOUNT60P']
luckyDrawDiscount = [10,15,30,50,60]
luckyDrawRedreemCode = ""
luckyDrawClaimed = False

validIdList = []
validNameList = []
validPriceList = []

class ChatApplication:
    
    def __init__(self):
        self.window = Tk()
        self._setup_main_window()
        
    def run(self):
        self.window.mainloop()
        
    def _setup_main_window(self):
        self.window.title("Eternal Restaurant Chatbot")
        self.window.resizable(width=True, height=True)
        self.window.configure(width=1000, height=700, bg=BG_COLOR)
        
        #head label
        head_label = Label(self.window, bg=BG_COLOR, fg=TEXT_COLOR, text="Welcome To Eternal Restaurant Chatbot", font=FONT_BOLD, pady=10)
        head_label.place(relwidth=1)
        
        #tiny divider
        line = Label(self.window, width=550, bg=BG_GRAY)
        line.place(relwidth=1, rely=0.07, relheight=0.012)
        
        #text widget
        self.text_widget = Text(self.window, width=20, height=2, bg=BG_COLOR, fg=TEXT_COLOR, font=FONT, padx=5, pady=5)
        self.text_widget.place(relheight=0.745, relwidth=0.983, rely=0.08)
        self.text_widget.configure(cursor="arrow", state=DISABLED)
        
        begin_text = "Hello, Welcome to Eternal Restaurant. I'm Chatbot Eternal. Please Tell Me Your Name~"
        msg1 = f"{CHATBOT_NAME}: {begin_text}\n\n\n"
        self.text_widget.configure(cursor="arrow", state=NORMAL)
        self.text_widget.insert(END, msg1)
        self.text_widget.configure(cursor="arrow", state=DISABLED)
        
        #scroll bar
        scrollbar = Scrollbar(self.window)
        scrollbar.place(relheight=1, relx=0.983)
        scrollbar.configure(command=self.text_widget.yview)
        
        #bottom label
        bottom_label = Label(self.window, bg=BG_GRAY, height=80)
        bottom_label.place(relwidth=1, rely=0.825)
        
        #message entry box
        self.msg_entry = Entry(bottom_label, bg="#2C3E50", fg=TEXT_COLOR, font=FONT)
        self.msg_entry.place(relwidth=0.74, relheight=0.06, rely=0.008, relx=0.011)
        self.msg_entry.focus()
        self.msg_entry.bind("<Return>", self._on_enter_pressed)
        
        #send button
        send_button = Button(bottom_label, text="Send", font=FONT_BOLD, width=20, bg=BG_GRAY, command=lambda: self._on_enter_pressed(None))
        send_button.place(relx=0.77, rely=0.008, relheight=0.06, relwidth=0.22)
        
    def _on_enter_pressed(self, event):
        msg = self.msg_entry.get()
        global ordering
        if ordering == False:
            self._insert_message(msg)
        else:
            self._insert_order(msg)
        
    def _insert_message(self, msg):
        if not msg:
            return
        
        global count
        global userName
        
        if count == 1:
            userName = msg
        
        self.print_user_input(msg)
        
        if count == 1:
            count += 1
            msg2 = f"{CHATBOT_NAME}: Hi {userName}.\nWelcome to Eternal Restaurant.\n Here are the feature that we provided:\n1. Menu\n2. Food List\n3. Beverage List\n4. Set List\n5. How To Order\n6. Help\n7. Frequently Asked Questions\n8. Recommended Food\n9. Recommended Beverage\n10.Recommended Set\nYou can type the key words or type the number to know more.\n\n\n"
            self.text_widget.configure(state=NORMAL)
            self.text_widget.insert(END, msg2)
            self.text_widget.configure(state=DISABLED)
            return
        
        global features
        
        for i in range(1,11):
            if msg == str(i):
                msg = features[i-1]
                
        answerIndex = self.searchQuestion(self.ignoreSymbol(msg).upper())
        if answerIndex != -12:
            msg = self.searchAnswer(answerIndex)
            msg2 = f"{CHATBOT_NAME}: \n{msg}\n\n\n"
            self.text_widget.configure(state=NORMAL)
            self.text_widget.insert(END, msg2)
            self.text_widget.configure(state=DISABLED)

            self.text_widget.see(END)
    
    def print_user_input(self, msg):
        self.msg_entry.delete(0, END)
        msg1 = f"{userName}: {msg}\n"
        self.text_widget.configure(cursor="arrow", state=NORMAL)
        self.text_widget.insert(END, msg1)
        self.text_widget.configure(cursor="arrow", state=DISABLED)
    
    def _insert_order(self, msg, print_user_input = True):
        global ordering_progress
        global userInformation
        global userName, CHATBOT_NAME
        if print_user_input == True:
            self.print_user_input(msg)
        msg = self.ignoreSymbol(msg).upper()
        if ordering_progress == 1 and msg == "YES":
            self.displayOrder()
        elif ordering_progress == 2 and msg == "CONFIRM":
            if userInformation == True:
                msg = "Type 'YES' if use back the previous order information(Name, Address, Phone No): \n\n"
                self.print_msg(f"{CHATBOT_NAME}: {msg}")
                ordering_progress = 8
            else:
                ordering_progress = 3
                self._insert_order("", False)
        elif ordering_progress == 3:
            msg = "Type 'YES' if use back the current name: \n\n"
            self.print_msg(f"{CHATBOT_NAME}: {msg}")
            ordering_progress = 4
        elif ordering_progress == 4:
            if msg == "YES":
                ordering_progress = 6
                self._insert_order(userName, False)
            else:
                ordering_progress = 5
                self._insert_order("", False)
        elif ordering_progress == 5:
            msg = "Please enter your name: \n\n"
            self.print_msg(f"{CHATBOT_NAME}: {msg}")
            ordering_progress = 6
        elif ordering_progress == 6:
            userName = msg
            msg = "Please enter your address:\n\n"
            self.print_msg(f"{CHATBOT_NAME}: {msg}")
            ordering_progress = 7
        elif ordering_progress == 7:
            global validatingPhoneNo
            if validatingPhoneNo == 1:
                global userAddress
                userAddress = msg
                msg = "Please enter your Phone No exactly in format (01x-xxxxxxx):\n\n"
                self.print_msg(f"{CHATBOT_NAME}: {msg}")
                validatingPhoneNo = 2
            elif validatingPhoneNo == 2:
                success = self.search_pattern_in_string("[0]{1}[1]{1}[0-4,6-9]{1}-[0-9]{7,8}", msg)
                if success == True:
                    ordering_progress = 9
                    global userPhoneNo
                    userInformation = True
                    userPhoneNo = msg
                    validatingPhoneNo = 1
                    self._insert_order("", False)
                else:
                    msg = "Wrong Format! Please enter your Phone No exactly in format (01x-xxxxxxx):\n\n"
                    self.print_msg(f"{CHATBOT_NAME}: {msg}")
        elif ordering_progress == 8:
            if msg == "YES":
                ordering_progress = 9
                self._insert_order("", False)
            else:
                ordering_progress = 5
                self._insert_order("", False)
        elif ordering_progress == 9:
            msg = "Any requests? Ex:'No vegetable':"
            self.print_msg(f"{CHATBOT_NAME}: {msg}\n")
            ordering_progress = 10
        elif ordering_progress == 10:
            global request
            request = msg
            self.displayReceipt()
            global ordering
            ordering = False
            ordering_progress = 1
        else:
            ordering = False
            ordering_progress = 1
            msg = "Order Cancelled\n\n"
            self.print_msg(f"{CHATBOT_NAME}: {msg}")
            
    def searchQuestion(self, userInput):
        global dataset
        answerList = []
        i = 0
        if userInput in ["RECOMMENDED SET","SUGGEST SET","SUGGESTED SET","SUGGEST A SET FOR ME","RECOMMEND A SET FOR ME", "SET MENU"]:
            userInput = self.setSuggest().upper()
        if userInput in ["MENU", "DO YOU HAVE A MENU", "CAN YOU SHOW ME THE MENU", "CAN U SHOW ME THE MENU", "CAN I HAVE A MENU", "CAN I TAKE A LOOK THE MENU", "CAN I TAKE A LOOK ON YOUR MENU", "CAN SHOW ME MENU", "CAN SHOW ME THE MENU", "CAN I TAKE A LOOK MENU", "CAN I TAKE A LOOK ON MENU"]:
            msg2 = f"{CHATBOT_NAME}: Yes, which menu do you prefer?(Main Course/ Beverage/ Set Menu) : \n\n\n"
            self.text_widget.configure(state=NORMAL)
            self.text_widget.insert(END, msg2)
            self.text_widget.configure(state=DISABLED)
            self.text_widget.see(END)
            return -12
        while(i < len(dataSet)):
            upperQuestion = self.convertQuestionToUpper(dataSet['Question'][i])
            if(upperQuestion == userInput):
                answerList.append(i)
                i += 1
            else:
                i += 1
        if(i == len(dataSet) and len(answerList) == 0):
            if "ORDER" in userInput:
                global ordering
                ordering = True
                msg2 = f"{CHATBOT_NAME}: \n"
                self.text_widget.configure(state=NORMAL)
                self.text_widget.insert(END, msg2)
                self.text_widget.configure(state=DISABLED)
                self.text_widget.see(END)
                self.foodOrder(userInput)
                return -12
            elif "LUCKY DRAW" in userInput:
                global luckyDrawRedreemCode, luckyDrawClaimed, luckyDraw, luckyDrawDiscount
                if luckyDrawClaimed == False:
                    luckyDrawRedreemCode = self.ignoreSymbol(rd.sample(luckyDraw,1))
                    i = 0
                    for code in luckyDraw:
                        if code == luckyDrawRedreemCode:
                            discount = luckyDrawDiscount[i]
                        i += 1
                    msg = "Congratulation, You get %d Percent Discount. \nNoted*: The discount will auto apply when you make an order." %(discount)
                    self.print_msg(f"{CHATBOT_NAME}: {msg}")
                    luckyDrawClaimed = True
                else:
                    msg = "Sorry, One people only can redreem one times."
                    self.print_msg(f"{CHATBOT_NAME}: {msg}")
                return -12
            return -1
        return answerList

    def searchAnswer(self, answersIndex):
        answerIndex = 0
        if(answersIndex == -1):
            noAnswer = rd.sample(['Sorry, I have no answer for that', 'Sorry, I do not understand.', 'Sorry, I did not get that.', 'Huh, What? I not really understand yor question.'], 1)
            suggestFeature = ("\nMaybe you can try the feature that we provided:\n"+
                             "1. Menu\n2. Food List\n3. Beverage List\n4. Set List\n5. How To Order\n6. Help\n7. Frequently Asked Questions\n8. Recommended Food\n9. Recommended Beverage\n10.Recommended Set"+
                             "\nYou can type the key words or type the number to know more.\n")
            reply = self.ignoreSymbol(noAnswer) + suggestFeature
            return reply
        if(len(answersIndex) > 1):
            answerIndex = rd.sample(answersIndex , 1)
            answerIndex = int(self.ignoreSymbol(answerIndex))
        else:
            answerIndex = answersIndex[0]
        return dataSet['answer'][answerIndex]
    
    def convertQuestionToUpper(self, dataSetQuestion):
        convertedQuestion = dataSetQuestion.upper()
        return convertedQuestion
    
    def setSuggest(self):
        currentTime = dt.datetime.now()#get current time
        currentHours = currentTime.hour
        if(currentHours >= 8 and currentHours < 12):
            return("Breakfast set")
        elif(currentHours >= 12 and currentHours < 15):
            return("Lunch set")
        elif(currentHours >= 15 and currentHours < 18):
            return("Tea time set")
        elif(currentHours >= 18 and currentHours < 21):
            return("Dinner set")
        elif(currentHours >= 21 and currentHours < 23):
            return("Suffer set")
        else:
            return("ERROR-CLOSED")
    
    def ignoreSymbol(self, string):
        string = str(string)
        squareBracket = "[]''.?!"
        for char in squareBracket:
            string = string.replace(char,"")
        return string
    
    def foodOrder(self, userInput):
        userInputBackup = userInput
        foodId = ""
        beverageId = ""
        packageId = ""
        orderPrice = []
        userOrderListId = []
        userOrderListName = []
        checkAgain = True
        for foodIndex in range(1, 12):
            if(foodIndex < 10):
                foodId = "F00" + str(foodIndex)
            else:
                foodId = "F0" + str(foodIndex)
            while(checkAgain == True):
                if foodId in userInput:
                    orderPrice.append(foodPrice[int(foodIndex)-1])#foodIndex start from 1, so need to deduct 1 to get the correct index
                    userInput = userInput.replace(foodId, "", 1)
                    userOrderListId.append(foodId)
                    userOrderListName.append(foodMenu[foodIndex-1])
                else:
                    checkAgain = False
            checkAgain = True
        for beverageIndex in range(1, 7):
            beverageId = "B00" + str(beverageIndex)
            while(checkAgain == True):
                if beverageId in userInput:
                    orderPrice.append(beveragePrice[beverageIndex-1])
                    userInput = userInput.replace(beverageId, "", 1)
                    userOrderListId.append(beverageId)
                    userOrderListName.append(beverageMenu[beverageIndex-1])
                else:
                    checkAgain = False
            checkAgain = True
        for packageIndex in range(1, 10):
            packageId = "P00" + str(packageIndex)
            while(checkAgain == True):
                if packageId in userInput:
                    orderPrice.append(packagePrice[packageIndex-1])
                    userInput = userInput.replace(packageId, "", 1)
                    userOrderListId.append(packageId)
                    userOrderListName.append(packageMenu[packageIndex-1])
                else:
                    checkAgain = False
            checkAgain = True
        success = self.checkOrder(userInput, userOrderListId, userOrderListName, orderPrice)
        if(success == True):
            global validIdList, validNameList, validPriceList
            validIdList = []
            validNameList = []
            validPriceList = []
            validIdList = userOrderListId
            validNameList = userOrderListName
            validPriceList = orderPrice
            self.displayOrder()
        
    def checkOrder(self, orderList, userOrderListId, userOrderListName, userOrderListPrice):
        failed = False
        invalidSetList = self.checkSetOrder(userOrderListId)
        invalidIdList = invalidSetList['id']
        invalidNameList = invalidSetList['name']
        global validIdList, validNameList, validPriceList, ordering
        validIdList = []
        validNameList = []
        validPriceList = []
        response = "N"
        for i in range(0,len(userOrderListId)):
            if not userOrderListId[i] in invalidIdList:
                validIdList.append(userOrderListId[i])
                validNameList.append(userOrderListName[i])
                validPriceList.append(userOrderListPrice[i])
        orderList_array = orderList.split()
        orderIndex = orderList_array.index("ORDER")
        empty = ""
        for i in range(0, orderIndex + 1):
            orderList_array[i] = empty
        orderList = "".join(orderList_array)
        orderList = orderList.replace(":", "")
        orderListBackup = orderList
        orderList = orderList.replace(" ", "")
        orderList = orderList.replace(",", "")
        orderListBackup = orderListBackup.replace(",", " ")
        if(orderList != "" or len(invalidIdList) != 0):
            msg = "Inside the Order, I found that there are some error in the order list."
            self.print_msg(msg)
            if(orderList != ""):
                msg = "The food code that is not inside our menu : %s" %(orderListBackup)
                self.print_msg(msg)
            if(len(invalidIdList) != 0):
                msg = "The food that currently not available  :"
                self.print_msg(msg)
                msg = "%3s %7s %-49s" %("No.","Food Id","Food Name")
                for i in range(0, len(invalidIdList)):
                    msg = "%02d. %-7s %-49s" %(i+1,invalidIdList[i],invalidNameList[i])
                    self.print_msg(msg)
                self.print_msg("")
            if(len(validIdList)!=0):
                msg = "And this is the order that I detect: "
                self.print_msg(msg)
                msg = "%3s %7s %-49s" %("No.","Food Id","Food Name")
                for i in range(0,len(validIdList)):
                    msg = "%02d. %-7s %-49s" %(i+1,validIdList[i],validNameList[i])
                    self.print_msg(msg)
                self.print_msg("")
                msg = "Did you want to use this list to continue payment? 'YES' for continue : \n\n"
                self.print_msg(msg)
                return False
            else:
                msg = "Sorry, there didn't have any food code that I able to detect."
                self.print_msg(msg)
                msg = "Order Cancelled\n\n"
                self.print_msg(msg)
                ordering = False
                return False
        return True
    
    def checkSetOrder(self, orderList):
        invalidSetId = []
        invalidSetName = []
        #   variable needed
        invalidSet_includeName = []
        currentTime = dt.datetime.now()#get current time
        currentHours = currentTime.hour
        setList = []
        checkAgain = True
        #   change orderList to string format (because need to use .replace function)
        orderList = "".join(orderList)
        for packageIndex in range(1, 10):
            packageId = "P00" + str(packageIndex)
            while(checkAgain == True):
                if packageId in orderList:
                    orderList = orderList.replace(packageId, "", 1)
                    setList.append(packageId)
                else:
                    checkAgain = False
            checkAgain = True
            invalidSet = []

        #   change setList to string format (because need to use .replace function)    
        setList = "".join(setList)
    #   check time
        if(currentHours < 8 or currentHours >= 12):
            for packageIndex in range(1, 3):
                packageId = "P00" + str(packageIndex)
                while(checkAgain == True):
                    if packageId in setList:
                        setList = setList.replace(packageId, "", 1)
                        invalidSet.append(packageId)
                    else:
                        checkAgain = False
                checkAgain = True
        if(currentHours < 12 or currentHours >= 15):
            for packageIndex in range(3, 5):
                packageId = "P00" + str(packageIndex)
                while(checkAgain == True):
                    if packageId in setList:
                        setList = setList.replace(packageId, "", 1)
                        invalidSet.append(packageId)
                    else:
                        checkAgain = False
                checkAgain = True
        if(currentHours < 15 or currentHours >= 18):
            for packageIndex in range(5, 7):
                packageId = "P00" + str(packageIndex)
                while(checkAgain == True):
                    if packageId in setList:
                        setList = setList.replace(packageId, "", 1)
                        invalidSet.append(packageId)
                    else:
                        checkAgain = False
                checkAgain = True
        if(currentHours < 18 or currentHours >= 21):
            for packageIndex in range(7, 9):
                packageId = "P00" + str(packageIndex)
                while(checkAgain == True):
                    if packageId in setList:
                        setList = setList.replace(packageId, "", 1)
                        invalidSet.append(packageId)
                    else:
                        checkAgain = False
                checkAgain = True
        if(currentHours < 21 or currentHours >= 23):
            for packageIndex in range(9, 10):
                packageId = "P00" + str(packageIndex)
                while(checkAgain == True):
                    if packageId in setList:
                        setList = setList.replace(packageId, "", 1)
                        invalidSet.append(packageId)
                    else:
                        checkAgain = False
                checkAgain = True

        #   change invalidSet to string format (because need to use .replace function)
        invalidSet = "".join(invalidSet)
        #set the invalidSetId and invalidSetName
        for packageIndex in range(1, 10):
            packageId = "P00" + str(packageIndex)
            while(checkAgain == True):
                if packageId in invalidSet:
                    invalidSet = invalidSet.replace(packageId, "", 1)
                    invalidSetId.append(packageId)
                    invalidSetName.append(packageMenu[packageIndex-1])
                else:
                    checkAgain = False
            checkAgain = True
        fields = {
            'id': invalidSetId,
            'name': invalidSetName,
        }
        return fields

    def calTotalPrice(self, orderList):
        totalPrice = 0.00
        for order in orderList:
            totalPrice += order
        return totalPrice
    
    def displayOrder(self):
        global validIdList, validNameList, validPriceList, ordering_progress, luckyDrawRedreemCode, luckyDraw, luckyDrawDiscount
        totalPrice = float(self.calTotalPrice(validPriceList))
        discountPercentage = 0
        discountPrice = 0
        redreem = False
        if(luckyDrawRedreemCode != ""):
            i = 0
            for code in luckyDraw:
                if code == luckyDrawRedreemCode:
                    discount = luckyDrawDiscount[i]
                i += 1
            backUpTotalPrice = totalPrice
            totalPrice = round(float(totalPrice * (100-discount)/100), 1)
            discountPrice = backUpTotalPrice-totalPrice
            redreem = True
        msg = "Double Check on Your Order:"
        self.print_msg(msg)
        i = 1
        msg = "%3s %7s %-79s %s" %("No.","Food Id","Food Name","Food Price")
        self.print_msg(msg)
        for price in validPriceList:
            msg = "%02d.  %-7s %-70s  RM%-.2f" %(i,validIdList[i-1],validNameList[i-1],float(price))
            self.print_msg(msg)
            i += 1
        if redreem == True:
            msg = "%98s %s RM%.2f" %("Discount Amount"," = RM",discountPrice)
            self.print_msg(msg)
        msg = "%98s %s RM%.2f" %("Total Price","=",totalPrice)
        self.print_msg(msg)
        msg = "If confirm payment please type 'Confirm' : \n\n"
        self.print_msg(msg)
        ordering_progress = 2

    def displayReceipt(self):
        global request, validIdList, validNameList, validPriceList, userName, userPhoneNo, userAddress, luckyDrawRedreemCode, luckyDraw, luckyDrawDiscount
        currentTime = dt.datetime.now()
        totalPrice = float(self.calTotalPrice(validPriceList))
        discountPercentage = 0
        discountPrice = 0
        redreem = False
        if(luckyDrawRedreemCode != ""):
            i = 0
            for code in luckyDraw:
                if code == luckyDrawRedreemCode:
                    discount = luckyDrawDiscount[i]
                i += 1
            luckyDrawRedreemCode = ""
            backUpTotalPrice = totalPrice
            totalPrice = round(float(totalPrice * (100-discount)/100), 1)
            discountPrice = backUpTotalPrice-totalPrice
            redreem = True
        date_string = currentTime.strftime("%d/%m/%Y")
        time_string = currentTime.strftime("%H:%M:%S")
        msg = "########################################################################"
        self.print_msg(msg)
        msg = "%67s" %("Eternal Restaurant Receipt")
        self.print_msg(msg)
        msg = "%s : %-12s   %70s: %-10s" %("Date",date_string,"Time",time_string)
        self.print_msg(msg)
        msg = "%s %-20s %55s  %12s" %("Name:",userName,"Phone No:", userPhoneNo)
        self.print_msg(msg)
        msg = "Address: %s" %(userAddress)
        self.print_msg(msg)
        msg = "%3s %7s %-79s %s" %("No.","Food Id","Food Name","Food Price")
        self.print_msg(msg)
        i = 1
        for price in validPriceList:
            msg = "%02d.  %-7s  %-70s  RM%-.2f" %(i,validIdList[i-1],validNameList[i-1],float(price))
            self.print_msg(msg)
            i += 1
        msg = "Requests Note: %s" %(request)
        self.print_msg(msg)
        if redreem == True:
            msg = "%98s %s RM%.2f" %("Discount Amount","=",discountPrice)
            self.print_msg(msg)
        msg = "%98s %s RM%.2f" %("Total Amount Paid","=",totalPrice)
        self.print_msg(msg)
        msg = "########################################################################"
        self.print_msg(msg)
        msg = "Thank You Very Much!!!\nHave A Nice Days~\n\n"
        self.print_msg(msg)

    def search_pattern_in_string(self, pattern, string):
        search = re.compile(pattern).search(string)
        if not search:
            return False
        else:
            if search.group() == string:
                return True
            else:
                return False
    
    def print_msg(self, msg):
        msg = f"{msg}\n"
        self.text_widget.configure(state=NORMAL)
        self.text_widget.insert(END, msg)
        self.text_widget.configure(state=DISABLED)
        self.text_widget.see(END)
        
if __name__ == "__main__":
    app = ChatApplication()
    app.run()
```
